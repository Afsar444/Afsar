```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/kumarkarna2/EDA_Notebooks/main/brain_stroke.csv')
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df)
plt.show()
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df, hue='gender')
plt.show()
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df,color='purple')
plt.show()
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df, hue='gender',style='work_type')
plt.show()
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df, hue='gender',style='work_type',alpha=0.3)
plt.show()
```

```python
sns.scatterplot(x='age',y='avg_glucose_level',data=df, hue='gender',style='work_type',alpha=0.3)
plt.legend(bbox_to_anchor=(1.2, 1), loc='upper left')
plt.show()
```

```python
sns.barplot(x='work_type',y='bmi',data=df)
```

```python
sns.barplot(x='work_type',y='bmi',hue='gender',data=df)
plt.show()
```

```python
sns.barplot(x="bmi", y="work_type", data=df, orient="h")
```

```python
sns.barplot(x='gender',y='hypertension',data=df,hue='smoking_status')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.title("Hypertension % in Gender by Smoking Status")
plt.xlabel('Gender')
plt.ylabel('Hypertension')
plt.show()
```

```python
df1=df.groupby(['gender','work_type'],as_index=False)['avg_glucose_level'].sum()
```

```python
df1
```

```python
sns.barplot(x='gender',y='avg_glucose_level',hue='work_type',data=df1)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
```

```python
sns.catplot(x="hypertension", y="avg_glucose_level", kind="violin", data=df)
plt.title("Distribution of avg_glucose_level by hypertension")
```

```python
sns.catplot(x="hypertension", y="avg_glucose_level", kind="box", data=df)
plt.title("Distribution of avg_glucose_level by hypertension")
```

```python
sns.catplot(
    x="hypertension",
    y="avg_glucose_level",
    kind="box",
    data=df,
    col="stroke",
    row="gender",
)
```

```python
sns.histplot(data=df, x="avg_glucose_level",bins=10)
```

```python
sns.displot(data=df, x="avg_glucose_level", kind="hist", hue="stroke", alpha=0.3)
```

```python
sns.displot(data=df, x="avg_glucose_level", kind="kde", hue="gender", alpha=0.5)
```

```python
sns.jointplot(data=df, x="avg_glucose_level", y="age")
```

```python
sns.pairplot(data=df)
```
